```python
import numpy as np
import matplotlib.pyplot as plt

impulse = np.zeros(10)
impulse[0] = 1

plt.stem(impulse)
plt.title("Unit Impulse")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python


step = np.ones(10)

plt.stem(step)
plt.title("Unit Step")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
n = np.arange(10)
ramp = n  # or ramp = np.arange(50)

plt.stem(n, ramp)
plt.title("Unit Ramp Signal")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
n = np.arange(30)
a = 0.9
real_exp = a ** n  # Real exponential decay

plt.stem(n, real_exp)
plt.title("Real Exponential Signal (a = 0.9)")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
import thinkdsp

# Create a sine signal of 440 Hz
sin_sig = thinkdsp.SinSignal(freq=440, amp=1.0)

# Generate a wave of duration 0.01 sec and sampling rate 44100 Hz
wave = sin_sig.make_wave(duration=0.01, framerate=44100)

# Plot the waveform
wave.plot()
plt.title("Sinusoidal Signal (440 Hz)")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.show()
```

```python
n = np.arange(30)
omega = np.pi / 4  # Frequency in radians/sample
complex_exp = np.exp(1j * omega * n)

plt.figure(figsize=(10, 4))

plt.stem(n, complex_exp.real, linefmt='b-', markerfmt='bo', basefmt='k-', label='Real Part')
plt.stem(n, complex_exp.imag, linefmt='r-', markerfmt='ro', basefmt='k-', label='Imaginary Part')
plt.title("Complex Exponential Signal (e^(jn))")
plt.xlabel("n")
plt.ylabel("Amplitude")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
